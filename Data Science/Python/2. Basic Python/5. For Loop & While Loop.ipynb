# Reverse the tuple element with using slicing and without using slicing (only loop)

# Dictionary

```python
d = {"a" : "fsds", "b" : "fsds", "c" : [1,2,3,4,5,6,7], "d":(4,5,6,7), "e":"sudh"}
```

```python
for i in d:
    print(i)
```

```python
for i in d.keys():
    print(i)
```
